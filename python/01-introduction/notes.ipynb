## Compreensão de Lista (List Comprehension)

```python
lista_1 =[x for x in range(5)]
lista_2 =[x ** 2 for x in range(5)]
lista_3 =[x ** 3 for x in range(5)]

print(lista_1,lista_2,lista_3)
```

```python
# Pegar os pares:
pares = [n for n in range(11) if n % 2 == 0]
impares = [n for n in range(11) if n % 2 == 1]

print(pares, impares)
```

```python
pessoas = [' Ana ', 'manuela', 'FELipe', 'pEdRO']

pessoas_normalizadas = [p.strip().capitalize() for p in pessoas]

print(pessoas_normalizadas)
```

## Classes e Objetos

```python
class JogadorLoteria:
    def __init__(self, nome):
        self.nome = nome
        self.numeros = (13,4,52,23,67,82)
    
    def total(self):
        return sum(self.numeros)

jogador_1 = JogadorLoteria("Pedro")

print(jogador_1.nome, jogador_1.numeros)

jogador_2 = JogadorLoteria("Pedro")

jogador_1 == jogador_2
```

```python
class Funcionario:
    aumento = 1.04

    def __init__(self, nome, salario):
        self.nome = nome
        self.salario = salario

    def dados(self,):
        return {'nome': self.nome, 'salario': self.salario}
    
    def aplicar_aumento(self,):
        self.salario = self.salario * self.aumento

    @classmethod
    def definir_novo_aumento(cls, novo_aumento):
        cls.aumento = novo_aumento
    
    @staticmethod
    def dia_util(dia):
        # segunda-feira: 0
        # domingo: 6
        # if dia.weekday() == 5 or dia.weekday() == 6:
        #     return False
        # return True
        return not(dia.weekday() == 5 or dia.weekday() == 6)


stanley = Funcionario('stanley', 3000.0)
print(stanley.dados())

class Admin(Funcionario):
    def __init__(self, nome, salario):
        super().__init__(nome, salario)
    
    def atualizar_dados(self, nome):
        self.nome = nome
        return self.dados()

dwight = Admin('dwight',5000)

print(dwight.dados())

dwight.atualizar_dados(nome='Dwight')

print(dwight.dados())
```

```python
stanley.aplicar_aumento()
stanley.dados()
```

```python
Funcionario.definir_novo_aumento(1.5) # aumentou 50%

novo_stanley = Funcionario(nome='stanley', salario=3000.0)

print(novo_stanley.dados())

novo_stanley.aplicar_aumento()

print(novo_stanley.dados())
```

## *args e **kwargs

```python
def soma(*args):
    return sum(args)

print(soma(1,2,4,5,6,7,45,6))
```

```python
def metodo_kwargs(*args, **kwargs):
    print(args)
    print(kwargs)

metodo_kwargs(3, 'saa',4, 'qualquer', nome='salumao', idade=34)
```
