## Compreensão de Lista (List Comprehension)

```python
lista_1 =[x for x in range(5)]
lista_2 =[x ** 2 for x in range(5)]
lista_3 =[x ** 3 for x in range(5)]

print(lista_1,lista_2,lista_3)
```

```python
# Pegar os pares:
pares = [n for n in range(11) if n % 2 == 0]
impares = [n for n in range(11) if n % 2 == 1]

print(pares, impares)
```

```python
pessoas = [' Ana ', 'manuela', 'FELipe', 'pEdRO']

pessoas_normalizadas = [p.strip().capitalize() for p in pessoas]

print(pessoas_normalizadas)
```
