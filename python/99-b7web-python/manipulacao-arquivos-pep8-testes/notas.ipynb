### Biblioteca pathlib

```python
import pathlib

# cria diretório
teste_folder = pathlib.Path("teste")

if teste_folder.exists() == False:
    teste_folder.mkdir() # Cria a pasta
    print("Pasta criada com sucesso!")

time.sleep(10)

if teste_folder.exists():
    teste_folder.rmdir()
    print("Pasta excluída com sucesso")
```
