```python
# Exemplo ruim

class Animal:

    def __init__(self, nome):
        self.__nome = nome

    def get_nome(self):
        return self.__nome

    def faz_som(self):
        if self.__nome == "Cachorro":
            print("Au Au")
        if self.__nome == "Gato":
            print("Miau")
```

## Princípio da Inversão de Dependências - D

```python
import abc

class Animal(abc.ABC):

    @abc.abstractmethod
    def faz_som(self):
        raise NotImplementedError()

class Cachorro(Animal):

    def faz_som(self):
        self.latir()

    def latir(self):
        print("Au Au")

class Gato(Animal):

    def faz_som(self):
        self.miar()

    def miar(self):
        print("Miau")
```

```python
# Exemplo ruim:
class Dono:

    def passear(self, cachorro: Cachorro):
        cachorro.latir()
```

```python
# Exemplo bom:
class Dono:

  def passear(self, animal: Animal):
        animal.faz_som()
```

```python
bicho = Cachorro()

claudio = Dono()

claudio.passear(bicho)

#xamego_2 = Animal()
```
