## Variáveis
Os principais tipos de variáveis em python são:
- int (inteiro)
- float (real)
- bool (True ou False)
- str (string, ou texto)
Segue abaixo alguns exemplos:

```python
# Variáveis: definição e tipos:
uma_string = "Salumão"
um_inteiro = 34
um_float = 1.68
um_boolean = True

# Imprimindo o valor e tipo:
print(uma_string)
print(type(uma_string))
```

```python
# Realizando casting
print(type(int(um_float)))
print(type(str(um_float)))
```

```python
# Operações aritméticas
resultado_1 = 7+5 # 12
resultado_2 = um_inteiro - resultado_1 # 22
resultado_3 = resultado_2 * 3 # 66
resultado_4 = resultado_3 / 4 # 16.5

# saída:
print(f"Resultado da soma = {resultado_1}")
print(f"Resultado da subtração = {resultado_2}")
print(f"Resultado da multiplicação = {resultado_3}")
print(f"Resultado da divisão = {resultado_4}")

# Potenciação e resto da Divisão
resultado_5 = 2 ** 4 # 16
resultado_6 = 13 % 4 # 3*4==12 e sobra 1 para 13

print(f"Resultado da potenciação = {resultado_5}")
print(f"Resultado do resto da divisão = {resultado_6}")

# Incremento e Decremento
um_inteiro -= 3
print("Resultado: " + str(um_inteiro)) # 34 - 3 = 31
um_inteiro += 6
print("Resultado: " + str(um_inteiro)) # 31 - 6 = 37
```

## Operadores relacionais
Segue abaixo os exemplos de aplicação:

```python
# Operadores de igualdade:
var_1 = 1
var_2 = 2
var_3 = 3
print(1 == var_1) # True
print(2 == var_1) # False

equal = (var_1 == var_2) # False
not_equal = (var_1 != var_2) # True

print("Um é igual a dois? " + str(equal))
print("Um é diferente de dois? " + str(not_equal))
```

## Strings

String armazenam uma cadeia de caracteres e existem diversas operações. A seguir teremos alguns exemplos comentados em código.

```python
# Concatenação
programacao = "Programação"
oo = "OO"
programacao_oo = programacao + " " + oo
print(programacao_oo)

# Tamanho
print(len(programacao_oo))

# Indexação e Substrings
frase = "Python é muito divertido!"
## Indexação começa com 0
print(frase[3]) # h
## -1 pega a última
print(frase[-1]) # !
## -2 pega a penúltima
print(frase[-2]) # n

# string[start:end] - intervalo fechado
print(frase[:4]) # do início até posição 4 (não incluso)
print(frase[10:]) # do índice 10 até final
print(frase[5:10]) # do índice 5 até o 10 (não incluso)
print(frase[:]) # imprime do começo ao fim
print(frase[::-1]) # inverte a string

print("Olá"*3) # repete olá 3 vezes

programacao_python = "Programação em Python"
print("Python" in programacao_python)
print("python" in programacao_python)
print("abacate" in programacao_python)

# Escaping
print("Estou estudando \n" + programacao_python) #\n 

# Maiúscula, minúscula, captalizado
print(programacao_python.lower())
print(programacao_python.upper())
print("aula de python".capitalize())

# Formatação

nome = "Salumão"
numero_inteiro = 7
numero_decimal = 7.584

print("Olá, meu nome é %s! Tenho %d anos e %d reais" % (nome,numero_inteiro, numero_decimal))
print("Olá, meu nome é %s! Tenho %d anos e %f reais" % (nome,numero_inteiro, numero_decimal))
print("Olá, meu nome é %s! Tenho %d anos e %.2f reais" % (nome,numero_inteiro, numero_decimal))
```

# Coleções

Coleções são estruturas de dados compostas. Exemplo, lista, tuplas e dicionários. Segue abaixo alguns exemplos:

```python
# Listas
numeros = [1, 10, 100, 1000]
print(numeros)
print(numeros[2]) # 100

# Operações
# inlcuir 2 itens na lista
numeros += [10000, 100000]
print(numeros)
# incluir 1 item usando append
numeros.append(1000000)
print(numeros)
# substituir os itens nos índices 1 e 2 por 7
numeros[1:3] = [7]
print(numeros)
# remove os itens nos índices 1 e 2
numeros[1:3] = []
print(numeros)
# tamanho da lista
print(len(numeros))
```

```python
# Tupla
naipes = ('copas', 'ouros', 'espadas', 'paus')
print(naipes)
```

```python
# Dicionário
notas = {"Ana": 8, "Maria": 5, "Thais": 10}
print(notas)
print(notas["Thais"]) # 10

# Incluir novo item
notas["Zaira"] = 9
print(notas)

# remove elemento
del notas["Thais"]
print(notas)

# checagem
print("Maria" in notas)
```

# Condições


```python
nome = "Ana"
idade = 25

if nome == "Ana":
    print("Passou!")

if nome == "Ana" or idade == 17:
    print("Passou de novo!")
    print("%s tem %d anos" % (nome, idade))
```

```python
dias = ["sábado", "domingo"]

if len(dias) == 0:
    print("Lista vazia")
elif len(dias) == 1:
    print("Só um dia")
else:
    print("Tamanho %d" % len(dias))
```

# Loops

Estrutura de repetição.

- for

```python
for i in range(6):
    print(i,end=" ")
print("\n")
for i in range(10, 0, -1):
    print(i, end=" ")
print("\n")
pares = [2, 4, 6, 8]
for i in range(len(pares)):
    print("Índice %d - valor %d " % (i, pares[i]))

hello_world = "Hello, World!"
for ch in hello_world:
    print(ch)

length = 0
for ch in hello_world:
    length += 1
print("Número de caracteres: %d" % length)
print("Número de caracteres: %d" % len(hello_world))
```

- while

```python
numero = 1 
while numero <= 7:
    print(numero)
    numero += 1

quadrado = 1
while quadrado <= 24:
    quadrado = numero ** 2
    print("Número %d - Quadrado %d" % (numero, quadrado))
    numero += 1
print("Ao sair do while, número tem o valor de %d" % numero)
```

- break

```python
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break

naipes = ('copas', 'ouros', 'espadas', 'paus')
lista_naipes = list(naipes) # conversão de tupla para lista
while True:
    # pop() - retira e retorna o último item da lista
    elemento = lista_naipes.pop() # paus - espadas
    print(elemento)
    if elemento == "espadas":
        break
```

- continue

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```
