## Variáveis
Os principais tipos de variáveis em python são:
- int (inteiro)
- float (real)
- bool (True ou False)
- str (string, ou texto)
Segue abaixo alguns exemplos:

```python
# Variáveis: definição e tipos:
uma_string = "Salumão"
um_inteiro = 34
um_float = 1.68
um_boolean = True

# Imprimindo o valor e tipo:
print(uma_string)
print(type(uma_string))
```

```python
# Realizando casting
print(type(int(um_float)))
print(type(str(um_float)))
```

```python
# Operações aritméticas
resultado_1 = 7+5 # 12
resultado_2 = um_inteiro - resultado_1 # 22
resultado_3 = resultado_2 * 3 # 66
resultado_4 = resultado_3 / 4 # 16.5

# saída:
print(f"Resultado da soma = {resultado_1}")
print(f"Resultado da subtração = {resultado_2}")
print(f"Resultado da multiplicação = {resultado_3}")
print(f"Resultado da divisão = {resultado_4}")

# Potenciação e resto da Divisão
resultado_5 = 2 ** 4 # 16
resultado_6 = 13 % 4 # 3*4==12 e sobra 1 para 13

print(f"Resultado da potenciação = {resultado_5}")
print(f"Resultado do resto da divisão = {resultado_6}")

# Incremento e Decremento
um_inteiro -= 3
print("Resultado: " + str(um_inteiro)) # 34 - 3 = 31
um_inteiro += 6
print("Resultado: " + str(um_inteiro)) # 31 - 6 = 37
```

## Operadores relacionais
Segue abaixo os exemplos de aplicação:

```python
# Operadores de igualdade:
var_1 = 1
var_2 = 2
var_3 = 3
print(1 == var_1) # True
print(2 == var_1) # False

equal = (var_1 == var_2) # False
not_equal = (var_1 != var_2) # True

print("Um é igual a dois? " + str(equal))
print("Um é diferente de dois? " + str(not_equal))
```
